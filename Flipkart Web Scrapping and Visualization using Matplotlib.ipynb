```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
a=urlopen("https://www.flipkart.com/laptops/~laptops-under-rs50000/pr?sid=6bo%2Cb5g")
html=a.read()
# html=a.close()
# print(html)
soup(html)
print(soup)
```

```python
soup_html=soup(html)
```

```python
filename="flipkart_laptp_list.csv"
f=open(filename,"wb")
f.write("Laptop_Name,Specification,MRP,Discount,Current_price,Rating \n".encode())

x=soup_html.findAll('div',{'class':'_3pLy-c row'})
linechange="\n"\


for i in x:
    for j in i.findAll('div',{'class':'_4rR01T'}):
        Laptop_Name=j.text.replace(",","-")
    
    for k in i.findAll('div',{'class':'fMghEO'}):
        Specification=k.text.replace(",","-")
        # print(o.text)
        
    for l in i.findAll('div',{'class':'_3I9_wc _27UcVY'}):
        MRP=l.text.replace(",","")
        # print(l.text)
        
    for t in i.findAll('div',{'class':'_3Ay6Sb'}):
        Discount=t.text
        # print(t.text)
    
    for n in i.findAll('div',{'class':'_30jeq3 _1_WHN1'}):
        Current_price=n.text.replace(",","")
        # print(n.text)
    
    for m in i.findAll('span',{'class':'_1lRcqv'}):
        Rating=m.text
    print(Laptop_Name+","+Specification+","+MRP+","+Discount+","+Current_price+","+Rating+"\n")
    
    f.write(f'{Laptop_Name},{Specification},{MRP},{Discount},{Current_price},{Rating}{linechange}'.encode())
f.close()
```

```python
import pandas as pd
Laptop_list=pd.read_csv("flipkart_laptp_list.csv",encoding="utf-8")
Laptop_list.index=Laptop_list.index +1
Laptop_list
```

```python
import matplotlib.pyplot as plt
```

## Replacing any white space present in the column name :

```python
Laptop_list.columns = [c.replace(' ','') for c in Laptop_list.columns]
```

## Filtering our MRP, Current_price, and Discount column

```python
Laptop_list['MRP'] = Laptop_list['MRP'].str.replace('₹','').astype(int)
Laptop_list['Current_price'] = Laptop_list['Current_price'].str.replace('₹','').astype(int)
Laptop_list['Discount'] = Laptop_list['Discount'].str.replace('% off','').astype(int)
Laptop_list.describe()
```

## Which laptop has highest rating ?

```python
plt.figure(figsize=(15,5))
plt.bar(Laptop_list['Laptop_Name'], Laptop_list['Rating'],color = 'blue',edgecolor='black')
plt.title("Laptops & Rating Graph", fontdict = {'fontsize':20,'fontweight':5,'color':'Red'})
plt.xlabel("Laptops Name", fontsize = 20)
plt.ylabel('Laptops Rating', fontsize = 20)
plt.xticks(rotation = 270)
plt.show()
```

## Which laptops have highest MRP and lowest MRP ?

```python
plt.bar(Laptop_list['Laptop_Name'], Laptop_list['MRP'],color = 'blue',edgecolor='black')
plt.title("Laptops & MRP Graph", fontdict = {'fontsize':20,'fontweight':5,'color':'Red'})
plt.xlabel("Laptops Name",fontsize = 20)
plt.ylabel('Laptops MRP', fontsize = 20)
plt.xticks(rotation = 270)
plt.show()
```

## Which laptop is giving the maximum discount ?

```python
plt.figure(figsize=(15,5))
plt.bar(Laptop_list['Laptop_Name'], Laptop_list['Discount'],color = 'blue',edgecolor='black')
plt.title("Laptops & Discount Graph", fontdict = {'fontsize':20,'fontweight':5,'color':'Red'})
plt.xlabel("Laptops Name",fontsize = 20)
plt.ylabel('Laptops Discount',fontsize = 20)
plt.xticks(rotation = 270)
plt.show()
```

## Which laptop would be the best buy considering the MRP and Discount of a Laptop_list data ?

```python
mrp_dis = Laptop_list.sort_values(by = ['MRP','Discount'], ascending = False)
```
